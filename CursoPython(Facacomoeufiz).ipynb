Nessa etapa, você deverá efetuar a leitura dos dados. Para isso importe a NumPy e use a função loadtxt. Use o link da url e o parâmetro usecols para pular a primeira coluna. Você pode usar np.arange para criar a sequência de números que representam as colunas.
```python
import numpy as np
url = "https://raw.githubusercontent.com/allanspadini/numpy/dados/citrus.csv"
dados = np.loadtxt(url,delimiter=",",usecols=np.arange(1,6,1),skiprows = 1)
```

Chegou a hora de você testar os conhecimentos desenvolvidos durante a aula. Continuando com o projeto das laranjas/toranjas agora você deve calcular o coeficiente ângular e o linear para a reta da laranja e para a reta da toranja. Use a fórmula de mínimos quadrados para encontrar cada um.

```python
#Vamos estipular uma regressão linear para laranja primeiro

n = np.size(peso_laranjas)
X = peso_laranjas
Y = diametro_laranjas

a = (n*np.sum(X*Y) - np.sum(X)*np.sum(Y))/(n*np.sum(X**2) - (np.sum(X)**2))
b = np.mean(Y) - a*np.mean(X)

y = a*X + b

#Criar um gráfico para laranjas

plt.plot(X,Y, "go--", color = "orange", markersize =1)
plt.plot(X,y, color = "green")
plt.xlabel("Peso")
plt.ylabel("Diâmetro")
```

```python
#Vamos estipular uma regressão linear para toranjas agora

n = np.size(peso_toranjas)
X = peso_toranjas
Y = diametro_toranjas

a = (n*np.sum(X*Y) - np.sum(X)*np.sum(Y))/(n*np.sum(X**2) - (np.sum(X)**2))
b = np.mean(Y) - a*np.mean(X)

y = a*X + b

#Criar um gráfico para laranjas

plt.plot(X,Y, "go--", color = "blue", markersize =1)
plt.plot(X,y, color = "orange")
plt.xlabel("Peso")
plt.ylabel("Diâmetro")
```

#Faça como eu fiz 4

Chegou a hora de você testar os conhecimentos desenvolvidos durante a aula. Continuando com o projeto das laranjas/toranjas, agora você deve calcular o coeficiente angular utilizando a geração de números aleatórios. Assuma que já conhece b e que este é igual a 17.

```python
b = 17
```

```python
#Primeiro vamos testar para laranjas e fixar uma seed
np.random.seed(84)
X = peso_laranjas
Y = diametro_laranjas
```

```python
#primeiro vamos gerar um array com valores aleatórios entre 0.1 e 0.9

coef_angular = np.random.uniform(0.0,30.0,100)

norma = np.array([])

for i in range(np.size(coef_angular)):
  norma = np.append(norma,np.linalg.norm(Y - (X*coef_angular[i]+b)))

#Agora podemos achar o valor estipulado do coeficiente atraves do menor valor do array
menor_norma = norma[0]
coef = coef_angular[0]
for i in range(np.size(coef_angular)):
  if norma[i] < menor_norma:
    menor_norma = norma[i]
    coef = coef_angular[i]

print(f"A norma com menor valor é: {menor_norma}\nE seu coeficiente é: {coef}")
```
